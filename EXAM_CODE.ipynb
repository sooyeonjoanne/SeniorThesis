```python
def importCSV(filename, *args):
  df2 = pd.read_csv(io.BytesIO(uploaded[filename]))
  arrofarrs = {}

  def removeNaN(arr):
    newList = []
    for i in range(len(arr)):
      if arr[i] == arr[i]:
        newList.append(arr[i])
    return newList

  for i in args:
    arrofarrs[i] = removeNaN(df2[i].tolist())

  return arrofarrs
```

# Normal Dist. Jarque-Bera Test

```python
data = dictionary["Allspecs"]

# Perform Jarque-Bera Test
jb_stat, p_value = stats.jarque_bera(data)

print("Jarque-Bera Statistic:", jb_stat)
print("P-value:", p_value)

# Interpretation of p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis - Data is not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Data is normally distributed.")
```

# Normal Dist. Shapiro-Wilk Test

```python
data = dictionary["Specs113exam"]

# Perform Shapiro-Wilk Test
statistic, p_value = stats.shapiro(data)

print("Shapiro-Wilk Statistic:", statistic)
print("P-value:", p_value)

# Interpretation of p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis - Data is not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Data is normally distributed.")
```

```python
data = dictionary["Trad113exam22"]

# Perform Shapiro-Wilk Test
statistic, p_value = stats.shapiro(data)

print("Shapiro-Wilk Statistic:", statistic)
print("P-value:", p_value)

# Interpretation of p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis - Data is not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Data is normally distributed.")
```

# Mann-Whitney U Test

```python
# All specs vs trad

group1 = dictionary["Allspecs"]
group2 = dictionary["Alltrad"]

# Perform the Mann-Whitney U Test
stat, p = mannwhitneyu(group1, group2)

#Output the results
print(f'U statistic: {stat}')
print(f'p-value: {p}')

#Decision rule for significance at alpha = 0.05
if p < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```

#Box Plot

```python
# All trad vs specs

data1 = dictionary["Alltrad"]
data2 = dictionary["Allspecs"]
data = [data1, data2]

medians = [np.median(d) for d in data]
means = [np.mean(d) for d in data]

fig, ax = plt.subplots()
box = ax.boxplot([data1,data2],showmeans=True, meanline=True)

info_text = (f"[Trad] Mean: {means[0]:.2f}%, Median: {medians[0]:.2f}%\n"
            f"[Specs] Mean: {means[1]:.2f}%, Median: {medians[1]:.2f}%")

ax.text(0.5, -0.15, info_text, transform=ax.transAxes, fontsize=10,
      ha='center', va='top', bbox=dict(boxstyle="round,pad=0.4", edgecolor="black", facecolor="lightgray"))

ax.set_ylim(25,102.5)
ax.set_yticks(range(30,105,10))

ax.set_xticklabels(["Traditional Course", "Specs Course"])

# Add title and labels
plt.title("Final Exam Scores for All Students")
plt.xlabel("Types of Grading")
plt.ylabel("Final Exam Score (%)")

# Show the plot
plt.show()

# Saving as raster image
fig.savefig("boxplot.pdf", format="pdf", bbox_inches="tight")
files.download("boxplot.pdf")
```
