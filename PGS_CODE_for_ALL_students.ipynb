```python
def importCSV(filename, *args):
  df2 = pd.read_csv(io.BytesIO(uploaded[filename]))
  arrofarrs = {}

  def removeNaN(arr):
    newList = []
    for i in range(len(arr)):
      if arr[i] == arr[i]:
        newList.append(arr[i])
    return newList

  for i in args:
    arrofarrs[i] = removeNaN(df2[i].tolist())

  return arrofarrs
```

# Normal Dist. Jarque-Bera Test

```python
data = dictionary["Outallspecs"]

# Perform Jarque-Bera Test
jb_stat, p_value = stats.jarque_bera(data)

print("Jarque-Bera Statistic:", jb_stat)
print("P-value:", p_value)

# Interpretation of p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis - Data is not normally distributed.")
else:
    print("Fail to reject the null hypothesis - Data is normally distributed.")
```

# Mann-Whitney U Test

```python
# Outlook - all specs vs trad

group1 = dictionary["Outallspecs"]
group2 = dictionary["Outalltrad"]

# Perform the Mann-Whitney U Test
stat, p = mannwhitneyu(group1, group2)

#Output the results
print(f'U statistic: {stat}')
print(f'p-value: {p}')

#Decision rule for significance at alpha = 0.05
if p < 0.05:
    print("There is a significant difference between the groups.")
else:
    print("There is no significant difference between the groups.")
```
